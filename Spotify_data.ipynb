```python
#insert output dataframe from the get_album_tracks function
def get_track_info(df):
    danceability = []
    energy = []
    key = []
    loudness = []
    speechiness = []
    acousticness = []
    instrumentalness = []
    liveness = []
    valence = []
    tempo = []
    row = 1
    for i in data['data.album.tracks.items.track.uri']:
        for x in sp.audio_features(tracks=[i]):
            danceability.append(x['danceability'])
            energy.append(x['energy'])
            key.append(x['key'])
            loudness.append(x['loudness'])
            speechiness.append(x['speechiness'])
            acousticness.append(x['acousticness'])
            instrumentalness.append(x['instrumentalness'])
            liveness.append(x['liveness'])
            valence.append(x['valence'])
            tempo.append(x['tempo'])
        print(i)
            
    df2 = pd.DataFrame({
    'danceability':danceability,
    'energy':energy,
    'key':key,
    'loudness':loudness,
    'speechiness':speechiness,
    'acousticness':acousticness,
    'instrumentalness':instrumentalness,
    'liveness':liveness,
    'valence':valence,
    'tempo':tempo})
    
    return df2
```

```python
def scrape_lyrics(artistname, songname):
    
    artistname2 = str(artistname.replace(' ','-').
                      replace("'","").
                      replace('.','').
                      replace("&","and").
                      replace(',','').
                      replace('ñ','n').
                      replace('É','e').
                      replace('Á','a').
                      replace('í','i')) if ' ' in artistname else str(artistname)

    songname = songname.strip()
    # 有的歌名是用全形英文符號，在genius網站的url表示法會不一樣...
    songname2 = str(songname.replace(" - Remix","").
                    replace(' ','-').
                    replace('.','').
                    replace(',','').
                    replace('$$','').
                    replace("'","").
                    replace("?","").
                    replace("!","").
                    replace("&","and").
                    replace('/','-').
                    replace('...','').
                    replace('…','').
                    replace(':-','').
                    replace('+',' ').
                    replace('ñ','n').                                    
                    replace('É','e').
                    replace('Á','a').replace('í','i')) if ' ' in songname else str(songname)
    if "(" and ")" in songname:
        songname = songname.strip()
        songname2 = str(songname.replace(" - Remix","").
                            replace(' ','-').
                            replace('.','').
                            replace(',','').
                            replace('$$','').
                            replace("'","").
                            replace("?","").
                            replace("!","").
                            replace("&","and").
                            replace('/','-').
                            replace('...','').
                            replace(':-','').
                            replace('+',' ').
                            replace('ñ','n').
                            replace('É','e').
                            replace('Á','a').
                            replace('í','i')) if ' ' in songname else str(songname)
        songname2 = re.sub("[\(\[].*?[\)\]]", "", songname2)
        songname2 = songname2[:-1]
    print(songname2)
    # print('https://genius.com/'+ artistname2 + '-' + songname2 + '-' + 'lyrics')

    page = requests.get('https://genius.com/'+ artistname2 + '-' + songname2 + '-' + 'lyrics')
    html = BeautifulSoup(page.text, 'html.parser')
    lyrics1 = html.find("div", class_="lyrics")
    lyrics2 = html.find("div", class_="Lyrics__Container-sc-1ynbvzw-6 jYfhrf")
    lyrics = lyrics2.get_text(separator=" ")
    if lyrics1 == lyrics2 == None:
        lyrics = None
    return lyrics
```
